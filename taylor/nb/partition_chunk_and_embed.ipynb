# Embed

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim

sentences = ["This is an example sentence", "Each sentence is converted", 'That is a happy person', 'That is a very happy person']

model = SentenceTransformer('sentence-transformers/sentence-t5-base')

```

```python
%%time
embeddings = model.encode(sentences)
print(embeddings)
print(cos_sim(embeddings[0], embeddings[1]))
print(embeddings[0] @ embeddings[1].T)
```

```python
%%time
embeddings = model.encode(sentences)
print(embeddings)
print(cos_sim(embeddings[0], embeddings[1]))
print(embeddings[0] @ embeddings[1].T)
```

```python
# https://github.com/FlagOpen/FlagEmbedding
model = SentenceTransformer('BAAI/bge-large-en-v1.5')
```

```python
%%time
embeddings = model.encode(sentences)
print(embeddings)
print(cos_sim(embeddings[0], embeddings[1]))
print(embeddings[0] @ embeddings[1].T)
```

```python
%%time
embeddings = model.encode(sentences)
print(embeddings)
print(cos_sim(embeddings[0], embeddings[1]))
print(embeddings[0] @ embeddings[1].T)
```

```python
%%time
embeddings_1 = model.encode(sentences[0], normalize_embeddings=True)
embeddings_2 = model.encode(sentences[1], normalize_embeddings=True)
print(embeddings_1)
print(embeddings_2)
print(cos_sim(embeddings_1, embeddings_2))
print(embeddings_1 @ embeddings_2.T)
```
